```python
import pandas as pd
lineup=pd.read_json (r'match_lineup.json')
df = pd.DataFrame(lineup)
matches = pd.read_csv('match.csv')
# print(lineup)
print(df)
```

### 1: Combination of the hero pool

```python
heroes= pd.read_csv('heroes.csv')
print(heroes.columns)
import itertools
x=itertools.combinations(heroes['Hero Name'], 2)
dict1={}
for i in x:
    dict1[i]={"Win":0, "Total":0}
dict1
```

```python
lineup1=lineup.transpose()
lineup1.head()
```

```python
y = matches['radiant_win'].apply(lambda win: 1 if win else 0)
# print(y)
classes = ['Dire Win', 'Radiant Win']
X= lineup1.merge(y, left_index=True, right_index=True)
X
```

Retrieve all the matches that radiant won the match

```python
radiant_win=X[X["radiant_win"]==1]
dire_win=X[X["radiant_win"]==0]
radiant_win.head()
# dire_win.head()
```

### 2. Hero Counter  Rate 

#####    Algorithm:
        loop through radiant_win:
            loop through rad_team:
                loop through dire_team:
                    if key in dict1 then win+=1, Total+=1
                    else Total+=1

```python
for index, row in radiant_win.iterrows():
    #look through every match
    rad = row[0]
#     print(rad)
    dire = row[1]
#     print(dire)
    
    for i in range (len(rad)):
        for j in range (len(dire)):
            key = (rad[i],dire[j])
#             print(key)
            if (key in dict1):
                try:
                    dict1[key]['Win']+=1
                    dict1[key]["Total"]+=1
                except:
                    pass
            else:
                try:
                    alt_key = (dire[j],rad[i])
                    dict1[alt_key]['Win']+=1
#                     print(alt_key)
                    dict1[alt_key]["Total"]+=1
                except:
                    pass
dict1
```

```python
hero_pair_counter_rate=pd.DataFrame.from_dict(dict1)
new=hero_pair_counter_rate.transpose()
new.head()
```

## 3. Top 10 Hero Counter Rate

```python
def counter_rate(x):
    new = x.copy()
    new['counter_rate'] = new['Win'] / new.Total
    return new
new1=counter_rate(new)

# Select hero pair that played more than 50 matches
more_than_50_matches = new1[new1['Total']>=50]

more_than_50_matches=more_than_50_matches.sort_values(by=['counter_rate'],ascending=False,)
top10 = more_than_50_matches.head(10)
top10
```

# 4.  Calculate Synergy for each team and their differences
#Adding the Synergy of each team in the dataframe

```python
def radiant(i):
#     print(i)
    synergy = 0
    i.sort()
    x=itertools.combinations(i, 2)
#     print(x)
    for j in x:
#         print(j)
        if (j[0] == 'Unknown' or j[1] == 'Unknown'):
            synergy+=0
        else:
            try:
                synergy+=dict1[j]['Win']/dict1[j]['Total']
            except:
                synergy+=dict1[j[1],j[0]]['Win']/dict1[j[1],j[0]]['Total']
    print(synergy)
    return synergy
    

X['synergy_radiant']=X['radiant'].apply(radiant)


def dire(i):
    synergy = 0
    i.sort()
    x=itertools.combinations(i, 2)
    for j in x:
        if (j[0] == 'Unknown' or j[1] == 'Unknown'):
            synergy+=0
        else:
            try:
                synergy+=dict1[j]['Win']/dict1[j]['Total']
            except:
                synergy+=dict1[j[1],j[0]]['Win']/dict1[j[1],j[0]]['Total']
    return synergy
    

X['synergy_dire']=X['dire'].apply(dire)
```

```python
X.head()
```

```python
def difference(x):
    new_df = x.copy()
    new_df['df_between_hero_synergy'] = new_df.synergy_radiant - new_df.synergy_dire
    return new_df

new_df=difference(X)
```
